```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

Train_dataset = pd.read_csv('C:\\TD\\Sem 1\\Courses\\Machine Learning\\train.csv')
X_train= Train_dataset.iloc[:, :-1].values
Y_train = Train_dataset.iloc[:, -1].values

Test_dataset = pd.read_csv('C:\\TD\\Sem 1\\Courses\\Machine Learning\\test.csv')
X_test = Test_dataset.iloc[:, 1:].values
```

```python
Y_train = Y_train.reshape(len(Y_train),1)
```

```python
print(Y_train)
```

```python
# Feature scaling for age in training set and test set
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,3:4] = sc.fit_transform(X_train[:, 3:4])
```

```python
X_test[:,2:3] = sc.transform(X_test[:, 2:3])
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
#Encoding for sex for training set
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train[:,1]= le.fit_transform(X_train[:,1])

```

```python
X_train
```

```python
#Encoding for Embarked(S/C/Q) for training set
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [2])], remainder='passthrough')
X_train = np.array(ct.fit_transform(X_train))
```

```python
X_train
```

```python
#Encoding for sex for test set
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_test[:,1]= le.fit_transform(X_test[:,1])
```

```python
print(X_test)
```

```python
#Encoding for Embarked(S/C/Q) for test set
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X_test = np.array(ct.fit_transform(X_test))
```

```python
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
print(classifier.predict([[0.0, 0.0, 1.0, 3, 1,sc.transform([[22]])]]))
```

```python
print(classifier.predict([[1.0, 0.0, 0.0, 1, 0,sc.transform([[38]])]]))
```

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```
